### Divide Data ###

```python
x = data.drop('Class', axis=1)
y = data['Class']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Scaling ###

```python
from sklearn.preprocessing import StandardScaler
import pickle

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

with open("../artifacts/scaler.pickle", "wb") as file:
    pickle.dump(scaler, file)
```

### Handle imbalanced dataset ###

```python
data['Class'].value_counts()
```

```python
import matplotlib.pyplot as plt
import numpy as np

counts = np.array([
    y_train.value_counts()[0],
    y_train.value_counts()[1]
])

labels = ['Legitimate', 'Fraud']

plt.figure()
plt.bar(labels, counts)
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Distribution of Training Classes")
plt.show()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_train_scaled_smote, y_train_smote = smote.fit_resample(x_train_scaled, y_train)
```

```python
import matplotlib.pyplot as plt
import numpy as np

counts = np.array([
    y_train_smote.value_counts()[0],
    y_train_smote.value_counts()[1]
])

labels = ['Legitimate', 'Fraud']

plt.figure()
plt.bar(labels, counts)
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Distribution of Training Classes")
plt.show()
```

### Model Training and Evaluation ###

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.metrics import classification_report

def training_scores(y_act, y_pred):
    acc = round(accuracy_score(y_act, y_pred), 3)
    pr = round(precision_score(y_act, y_pred), 3)
    rec = round(precision_score(y_act, y_pred), 3)
    f1 = round(f1_score(y_act, y_pred), 3)

    print(f'Training Scores:\n\tAccuracy = {acc}\n\tPrediction = {pr}\n\tRecall = {rec}\n\tF1-Score = {f1}')

    print("")
    print(classification_report(y_act, y_pred))
```

```python
def validation_score(y_act,y_pred):
    acc = round(accuracy_score(y_act, y_pred), 3)
    pr = round(precision_score(y_act, y_pred), 3)
    rec = round(recall_score(y_act, y_pred), 3)
    f1 = round(f1_score(y_act, y_pred), 3)

    print(f'Testing Scores:\n\tAccuray = {acc}\n\tPrecision = {pr}\n\tRecall = {rec}\n\tF1-Score = {f1}')
    
    print("")
    print(classification_report(y_act, y_pred))
```

#### Logistic regression ####

```python
lr = LogisticRegression()
lr.fit(x_train_scaled_smote,y_train_smote)
y_train_pred = lr.predict(x_train_scaled_smote)
y_test_pred = lr.predict(x_test_scaled)
training_scores(y_train_smote, y_train_pred)
validation_score(y_test, y_test_pred)
```

#### GaussianNB ####

```python
gnb = GaussianNB()
gnb.fit(x_train_scaled_smote,y_train_smote)
y_train_pred = gnb.predict(x_train_scaled_smote)
y_test_pred = gnb.predict(x_test_scaled)
training_scores(y_train_smote, y_train_pred)
validation_score(y_test, y_test_pred)
```

#### XGboost ####

```python
xgb = XGBClassifier()
xgb.fit(x_train_scaled_smote,y_train_smote)
y_train_pred = xgb.predict(x_train_scaled_smote)
y_test_pred = xgb.predict(x_test_scaled)
training_scores(y_train_smote, y_train_pred)
validation_score(y_test, y_test_pred)
```

```python
models = ['Logistic Regression', 'GaussianNB', 'XGBoost']

accuracies = [0.974, 0.976, 0.999]  
f1_scores = [0.107, 0.112, 0.82]   
precisions = [0.057, 0.06, 0.804]   
recalls = [0.918, 0.867, 0.837]     

x = range(len(models))

plt.figure(figsize=(10,6))

plt.plot(x, accuracies, marker='o', linestyle='-', color='blue', label='Accuracy')
plt.plot(x, f1_scores, marker='s', linestyle='--', color='orange', label='F1-Score')
plt.plot(x, precisions, marker='^', linestyle='-.', color='green', label='Precision')
plt.plot(x, recalls, marker='d', linestyle=':', color='red', label='Recall')

for i in x:
    plt.text(i, accuracies[i]+0.01, f"{accuracies[i]:.3f}", ha='center', fontweight='bold')
    plt.text(i, f1_scores[i]+0.01, f"{f1_scores[i]:.3f}", ha='center', fontweight='bold')
    plt.text(i, precisions[i]+0.01, f"{precisions[i]:.3f}", ha='center', fontweight='bold')
    plt.text(i, recalls[i]+0.01, f"{recalls[i]:.3f}", ha='center', fontweight='bold')

plt.xticks(x, models)
plt.ylim(0, 1.05)
plt.ylabel("Score")
plt.title("Test Metrics for Different Models")
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
